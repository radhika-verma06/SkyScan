```python
# 3. AUTO-FIX data.yaml for Colab Paths
import yaml
import os

data_config = {
    'path': '/content', # Base directory
    'train': 'train/images', # Relative to path
    'val': 'valid/images',   # Relative to path
    'nc': 4,
    'names': ['antenna', 'chimney', 'power-lines', 'wind-turbine']
}

with open('data.yaml', 'w') as f:
    yaml.dump(data_config, f)

print("✅ data.yaml configured for Google Colab GPU training.")
```

```python
# 4. Start High-Precision Training
from ultralytics import YOLO
import torch

# Double check GPU status
device = 0 if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

model = YOLO('yolov8s.pt') # Small model for high accuracy

results = model.train(
    data='data.yaml',
    epochs=50,
    imgsz=640,
    device=device,
    project='windmill_runs',
    name='high_precision_windmill',
    mixup=0.1,  # Improved augmentation
    mosaic=1.0,
    patience=10 # Early stopping if no improvement
)
```
